```python
df=df[["movie_id","title","overview","genres","keywords","cast","crew"]]
```

```python
df.dropna(inplace=True)
```

```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
df['genres']=df['genres'].apply(convert)
```

```python
df.head()
```

```python
df['keywords']=df['keywords'].apply(convert)
```

```python
#for cast
def convert2(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=5:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
df['cast']=df['cast'].apply(convert2)
```

```python
#for crew
def convert3(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
df['crew']=df['crew'].apply(convert3)
```

```python
df.head()
```

```python
df['overview']=df['overview'].apply(lambda x:x.split())
```

```python
df.head()
```

```python
df['genres']=df['genres'].apply(lambda x:[i.replace(" ","") for i in x])
df['keywords']=df['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
df['cast']=df['cast'].apply(lambda x:[i.replace(" ","") for i in x])
df['crew']=df['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
df.head()
```

```python
df['tags']=df['overview']+df['genres']+df['keywords']+df['cast']+df['crew']
```

```python
df.head()
```

```python
new_df=df[['movie_id','title','tags']]
```

```python
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'][0]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer= WordNetLemmatizer()
def lemma(text):
    y=[]
    for i in text.split():
        y.append(wordnet_lemmatizer.lemmatize(i))
    return " ".join(y)
```

```python
new_df['tags'].apply(lemma)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=1000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
new_df['tags'][1]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:8]
    recommended_movies = []
    for i in movies_list:
        recommended_movies.append(new_df.iloc[i[0]].title)
    return recommended_movies
```

```python
recommend('The Avengers')
```

```python
import pickle
pickle.dump(new_df, open('movies.pkl','wb'))
pickle.dump(similarity, open('similarity.pkl','wb'))
```
